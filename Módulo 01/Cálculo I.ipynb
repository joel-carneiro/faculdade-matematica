# Cálculo Diferencial e Integral

O Cálculo Diferencial e Integral é um ramo da matemática que se dedica ao estudo das funções e suas propriedades relacionadas às taxas de variação. O cálculo diferencial tem como objetivo determinar a taxa de variação instantânea de uma função em um ponto específico, enquanto o cálculo integral busca calcular a área sob uma curva, ou seja, a acumulação de valores de uma função ao longo de um intervalo.

Esses dois conceitos estão intimamente relacionados e são complementares, sendo que o cálculo diferencial é utilizado para encontrar a derivada de uma função, e o cálculo integral é utilizado para encontrar a área sob a curva dessa mesma função. O cálculo diferencial e integral tem aplicações em diversas áreas, como engenharia, física, economia, entre outras.

## 1.1 - Instânciando uma função

```python
def create_function(string):
    x = smp.Symbol('x')
    f = eval(string)
    
    return f

def create_graph(f):
    dom = np.arange(1, 100)
    img = [f.subs('x', k).evalf() for k in dom]
    
    plt.title(f)
    
    plt.plot(dom, img)
```

```python
f = create_function('x**2 / 2 - 2 * x + 18')
g = create_function('x**3 + smp.sin(x) - 18 * x**2')
h = create_function('smp.sin(x)')
i = create_function('smp.log(x, 10)')
```

```python
create_graph(f)
```

```python
create_graph(g)
```

```python
create_graph(h)
```

```python
create_graph(i)
```
